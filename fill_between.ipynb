```python
x = np.linspace(0, 10, 200)
```

```python
y1 = 2 * x + 1
```

```python
y2 = 3 * x + 1.2
```

```python
y_mean = 0.5 * x * np.cos(2*x) + 2.5 * x + 1.1
```

```python
y_mean
```

```python
fig, ax = plt.subplots()
```

```python
ax.fill_between(x, y1, y2)
```

```python
plt.show()
```

```python
ax.plot(x, y_mean)
```

```python
plt.show()
```

```python
ax.plot(x, y_mean, color='black')
```

```python
x = np.linspace(0, 10, 200)
```

```python
data_obj = {
    'x': x, 
    'y1': 2 * x + 1,
    'y2': 3 * x + 1.2,
    'mean': 0.5 * x * np.cos(2*x) + 2.5 * x + 1.1
}
```

```python
data_obj
```

```python
ax.fill_between('x', 'y1', 'y2', color='yellow', data=data_obj)
```

```python
plt.show()
```

```python
ax.plot('x', 'mean', color="white", data=data_obj)
```

```python
plt.show()
```
