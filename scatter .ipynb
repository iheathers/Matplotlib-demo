```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(1874)
```

```python
x, y, z = np.random.normal(0, 1, (3, 100))
```

```python
x
```

```python
x.shape
```

```python
t = np.arctan2(y,x)
```

```python
size = 50 * np.cos(2 * t)**2 + 10
```

```python
fig, axes = plt.subplots(ncols = 3)
```

```python
fig.tight_layout()
```

```python
plt.show()
```

```python
axes[0].scatter(x, y, marker='o', facecolor='white', edgecolor='black', s=80)
```

```python
plt.show()
```

```python
axes[1].scatter(x, y, s=size, marker='s', color='darkblue')
```

```python
plt.show()
```

```python
axes[2].scatter(x, y, c=z, s = size, cmap='gist_ncar')
```

```python
plt.show()
```
